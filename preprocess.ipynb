# creating json files

```python
def behaviour_generator(dataframe):
    data = []

    for index, row in dataframe.iterrows():
      entry = {}
      if row['type'] == 'Video':
        human = f"""<video>\n This tweet is by {row["inferred company"]} through username {row["username"]}, posted {row["date_difference"].days} days ago.
 The text content eclosed in double brackets is (( {row["content"]} )), duration is {row["duration"]} and views are {row["views"]}.
 Seeing the Video, text content, views and duration predict the number of likes."""
        gpt = str(row['likes'])
        entry = {
            "id": str(row['id']),
            "video": row['url'],
            "conversations": [
                {"from": "human", "value": human},
                {"from": "gpt", "value": gpt}
            ]
        }
        data.append(entry)
      elif row['type'] == 'Gif':
        human = f"""<video>\n This tweet is by {row["inferred company"]} through username {row["username"]}, posted {row["date_difference"].days} days ago.
 The text content eclosed in double brackets is (( {row["content"]} )).
 Seeing the Video and text content predict the number of likes."""
        gpt = str(row['likes'])
        entry = {
            "id": str(row['id']),
            "video": row['url'],
            "conversations": [
                {"from": "human", "value": human},
                {"from": "gpt", "value": gpt}
            ]
        }
        data.append(entry)
      else:
        human = f"""<image>\n This tweet is by {row["inferred company"]} through username {row["username"]}, posted {row["date_difference"].days} days ago.
 The text content eclosed in double brackets is (( {row["content"]} )).
 Seeing the Image and text content predict the number of likes."""
        gpt = str(row['likes'])
        entry = {
            "id": str(row['id']),
            "image": row['url'],
            "conversations": [
                {"from": "human", "value": human},
                {"from": "gpt", "value": gpt}
            ]
        }
        data.append(entry)

    return data

    # df_output = json.dumps(data, indent = 2)
    # with open(json_name, "w") as outfile:
    #   outfile.write(df_output)
```

```python
def content_generator(dataframe):
    data = []

    for index, row in dataframe.iterrows():
      entry = {}
      if row['type'] == 'Video':
        human = f"""<video>\n This tweet is by {row["inferred company"]} through username {row["username"]}, posted {row["date_difference"].days} days ago.
 The likes on tweet are {row["likes"]}, duration is {row["duration"]} and views are {row["views"]}.
 Seeing the Video, number of likes, views and duration generate the content."""
        gpt = row['content']
        entry = {
            "id": str(row['id']),
            "video": row['url'],
            "conversations": [
                {"from": "human", "value": human},
                {"from": "gpt", "value": gpt}
            ]
        }
        data.append(entry)
      elif row['type'] == 'Gif':
        human = f"""<video>\n This tweet is by {row["inferred company"]} through username {row["username"]}, posted {row["date_difference"].days} days ago.
  The likes on tweet are {row["likes"]}.
  Seeing the Video and number of likes generate the content."""
        gpt = row['content']
        entry = {
            "id": str(row['id']),
            "video": row['url'],
            "conversations": [
                {"from": "human", "value": human},
                {"from": "gpt", "value": gpt}
            ]
        }
        data.append(entry)
      else:
        human = f"""<image>\n This tweet is by {row["inferred company"]} through username {row["username"]}, posted {row["date_difference"].days} days ago.
  The likes on tweet are {row["likes"]}.
  Seeing the Image and number of likes generate the content."""
        gpt = row['content']
        entry = {
            "id": str(row['id']),
            "image": row['url'],
            "conversations": [
                {"from": "human", "value": human},
                {"from": "gpt", "value": gpt}
            ]
        }
        data.append(entry)
    return data
    # df_output = json.dumps(data, indent = 2)
    # with open(json_name, "w") as outfile:
    #   outfile.write(df_output)
```

```python
def create_json(data, json_name):
    data_output = json.dumps(data, indent = 2)
    with open(json_name, "w") as outfile:
      outfile.write(data_output)
```

```python
behaviour_data = behaviour_generator(df)
content_data = content_generator(df)
merged_data = behaviour_generator(df) + content_generator(df)
```
